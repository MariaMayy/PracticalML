### Attractor Lorenza
    x' = σ(y-x)
    y' = x(r-z)-y
    z' = x*y-b*z

```python
from numpy import linspace, array
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# constants
sigma = 10.0 
r = 28.0
beta = 8.0/3.0

# ----- Attractor Lorenza -----
def AttractorLorenza(t, vec):
    x, y, z = vec
    rX = sigma*(y - x)
    rY = r*x - y - x*z
    rZ = x*y - beta*z
    
    return array([rX, rY, rZ], float)

y0 = [0, 1, 0] # start point
# t_span = [0, 100]
solution = solve_ivp(AttractorLorenza, [0,100], y0, t_eval = linspace(0, 40, 5000))

fig = plt.figure()
ax = fig.gca(projection = '3d')

ax.plot(solution.y[0], solution.y[1], solution.y[2], 'red')
ax.set_xlabel("X", fontsize = 14, fontweight = "regular")
ax.set_ylabel("Y", fontsize = 14, fontweight = "regular")
ax.set_zlabel("Z", fontsize = 14, fontweight = "regular")
ax.set_title("Attractor Lorenza", fontsize = 14, loc = "center")

plt.show()
```
